```python
import pandas as pd
from apyori import apriori
```

```python
#encoding = "cp1252", encoding='utf-8', encoding='latin1'etc
base= pd.read_csv('/content/sample_data/paoemanteiga.csv', sep=';',encoding='cp1252', header=None)
base
```

```python
items = base.iloc[0, 1:].tolist()  # Obter nomes dos itens
transactions = base.iloc[1:].reset_index(drop=True)
```

```python
base.shape
base
```

```python
transactions
```

**Como estaremos treinando um modelo a priori, que recebe entradas em formato de lista, precisamos transformar o dataframe em uma lista de transações.**

```python
transaction_list = []
for index, row in transactions.iterrows():
    transaction = [items[i] for i in range(len(items)) if row[i+1] != 'Sim'] # mudando para selecionar os "Nao" (ausentes)
    transaction_list.append(transaction)

# O resultado é uma lista de transações onde cada transação contém apenas os itens ausentes!
print(transaction_list)
transacoes = transaction_list
```

```python
transacoes
```

```python
type(transacoes)
```

**Vamos chamar executar o algoritmo apriori e armazenar as regras obtidas**

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8)
saida = list(regras)
print(len(saida))
print(saida)
```
